```python
import matplotlib.pyplot
import numpy

sin_x = numpy.linspace(-5, 5, 1000)
sin_y = numpy.sin(sin_x)
cos_x = numpy.linspace(-5, 5, 1000)
cos_y = numpy.cos(cos_x)
tan_x = numpy.linspace(-5, 5, 1000)
tan_y = numpy.tan(tan_x)

matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.plot(sin_x, sin_y, cos_x, cos_y, tan_x, tan_y)
matplotlib.pyplot.title("")
matplotlib.pyplot.xlabel("")
matplotlib.pyplot.ylabel("")
matplotlib.pyplot.xticks(numpy.arange(-5, 5, 1))
matplotlib.pyplot.yticks(numpy.arange(-5, 5, 1))
matplotlib.pyplot.xlim()
matplotlib.pyplot.ylim(-2, 2)
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
import numpy

x = numpy.linspace(0, 5, 1000)
y = -1/5*x*numpy.cos(2*x*numpy.pi)

matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.title("(a) Example of continuous time signal")
matplotlib.pyplot.xlabel("times(s)")
matplotlib.pyplot.ylabel("")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

### (a) Example of continuous time signal
$ x(t) = -{1 \over 5} t \cos(2 \pi t) $ 

```python
import matplotlib.pyplot
import numpy

x = numpy.arange(0, 41, 1)
y = -1/40*x*numpy.cos(1/4*x*numpy.pi)

matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.scatter(x, y)
matplotlib.pyplot.title("(b) Example of a discrete-time signal")
matplotlib.pyplot.xlabel("sample index[n]")
matplotlib.pyplot.ylabel("")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

### (b) Example of a discrete-time signal
$ x[n] = -{1 \over 40} n \cos({1 \over 4} \pi n) $
