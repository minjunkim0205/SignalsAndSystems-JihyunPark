```python
import matplotlib.pyplot
import numpy

x = numpy.linspace(0, 5, 10000)
y = 1/5*x*numpy.cos(2*x*numpy.pi)

matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.title("(a) Example of continuous time signal")
matplotlib.pyplot.xlabel("times(s)")
matplotlib.pyplot.ylabel("")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
import numpy

x = numpy.arange(0, 40, 1)
y = 1/40*x*numpy.cos(1/4*x*numpy.pi)

matplotlib.pyplot.scatter(x, y)
matplotlib.pyplot.title("(b) Example of a discrete-time signal")
matplotlib.pyplot.xlabel("sample index[n]")
matplotlib.pyplot.ylabel("")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```
