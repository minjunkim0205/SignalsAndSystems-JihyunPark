# 실습 - 정현파 

### 파이썬을 이용해서 정현파 그리기

```python
import matplotlib.pyplot
import numpy

t = numpy.linspace(0, 3*numpy.pi, 11)
x = numpy.sin(t)

matplotlib.pyplot.title("Example of plot(t, x) : 11-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.plot(t, x)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
import numpy

t = numpy.linspace(0, 3*numpy.pi, 51)
x = numpy.sin(t)

matplotlib.pyplot.title("Example of plot(t, x) : 51-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.plot(t, x)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
import numpy

t = numpy.linspace(0, 3 * numpy.pi, 51)
x = numpy.sin(t)

ticks_t = numpy.linspace(0, 3 * numpy.pi, 7)
ticks_label = ["$0$", "$\\frac{1}{2}\pi$", "$\pi$",
               "$\\frac{3}{2}\pi$", "$2\pi$", "$\\frac{5}{2}\pi$", "$3\pi$"]

matplotlib.pyplot.title("Example of plot(t, x) : 51-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.xticks(ticks_t, ticks_label)
matplotlib.pyplot.plot(t, x)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
import numpy

t_s = 3*numpy.pi/25
n = numpy.arange(26)
x = numpy.sin(t_s*n)

matplotlib.pyplot.title("Example of stem(n, x) : 26-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (10, 4)
matplotlib.pyplot.xticks(ticks_t, ticks_label)
matplotlib.pyplot.xlabel("$n$")
matplotlib.pyplot.stem(n, x)
matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```
