## Week01-2 수업에서 손으로 그렸던 그래프 그리기

### Te01
$ x(t) = 20 \cos( 2 \pi (40) t -0.4 \pi ) $ 

```python
import matplotlib.pyplot
import numpy

x = numpy.linspace(-0.04, 0.04, 201)
y = 20*numpy.cos(2*numpy.pi*(40)*x-0.4*numpy.pi)

matplotlib.pyplot.title("Te01 : 201-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (10, 3)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.show()
```

### Te02-1
$ f_0 = 100 \\
x(t) = 5 \cos( 2 \pi f_0 t ) $

```python
import matplotlib.pyplot
import numpy

f_0 = 100
x = numpy.linspace(-0.02, 0.02, 201)
y = 5*numpy.cos(2*numpy.pi*f_0*x)

matplotlib.pyplot.title("Te02-1 : 201-points")
matplotlib.pyplot.rcParams["figure.figsize"] = (12, 2)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.show()
```
